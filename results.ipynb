```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
with open('results/dqn_rewards.pkl', 'rb') as f:
    dqn_rewards = pkl.load(f)
```

```python
with open('results/grd_rewards.pkl', 'rb') as f:
    grd_rewards = pkl.load(f)
```

```python
plt.plot(range(len(grd_rewards)), (grd_rewards), label='grd')
plt.plot(range(len(dqn_rewards)), (dqn_rewards), label='dqn')
plt.title('raw rewards')
plt.ylabel('rewards')
plt.xlabel('episodes')
plt.legend()

plt.show()
```

```python
plt.plot(range(len(grd_rewards)), np.cumsum(grd_rewards), label='grd')
plt.plot(range(len(dqn_rewards)), np.cumsum(dqn_rewards), label='dqn')
plt.ylabel('cumulative rewards')
plt.xlabel('episodes')
plt.title('cumulative rewards')
plt.legend()

plt.show()
```
